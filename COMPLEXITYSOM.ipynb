```
# COMPLEXITY WITHIN THE LIMIT
from minisom import MiniSom    
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns; sns.set_theme()
from sklearn.datasets import load_digits
from pylab import plot,axis,show,pcolor,colorbar,bone
data= pd.read_csv('/dataset_after_data_complexity.csv')
#print(data.info())
som = MiniSom(5, 5, 2, sigma=0.5, learning_rate=0.1)
X=data.iloc[:, [0,2]].values
labels=X
som.random_weights_init(X)
print("Initiating SOM.")
#som.train_random(X,100) 
som.train_batch(X,100) 
print("\n. SOM Processing Complete")
pcolor(som.distance_map().T) 
colorbar()
print(som.distance_map().T)
markers = ['o', '1', '^', 'h', 'X', 's', 'p', 'P', 'D', '*']
colors = ["r", "g", "b", "y", "c", (0,0.1,0.8), (1,0.5,0), (1,1,0.3), "m", (0.4,0.6,0)]
y=[]
for x in data.Game:
   y.append(x)
for x in range(0,10):
     print(x,markers[x])
pcolor(som.distance_map().T) 


j=0.0001
for i,x in enumerate(X):
    w=som.winner(x)
    plot(w[0]+j,
         w[1]+j,
        markers[y[i]-1],
        markeredgecolor=colors[y[i]-1],markerfacecolor='None',markersize=10,markeredgewidth=2)
    j=j+0.0009         
show()
pcolor(som.distance_map().T) 
colorbar()
j=0.001  
for i,x in enumerate(X):
    w=som.winner(x)
    if(  x[1]==1):
      plot(w[0]+j,
         w[1]+j,
        markers[0],
        markeredgecolor=colors[0],markerfacecolor='None',markersize=10,markeredgewidth=2)
      j=j+0.009
     
show()
pcolor(som.distance_map().T) 
colorbar()
j=0.001  
for i,x in enumerate(X):
    w=som.winner(x)
    if(  x[1]==2):
      plot(w[0]+j,
         w[1]+j,
        markers[1],
        markeredgecolor=colors[1],markerfacecolor='None',markersize=10,markeredgewidth=2)
      j=j+0.009
    
            
show()
pcolor(som.distance_map().T) 
colorbar()
j=0.001
 
  
for i,x in enumerate(X):
    w=som.winner(x)
    if(  x[1]==3):
      plot(w[0]+j,
         w[1]+j,
        markers[2],
        markeredgecolor=colors[2],markerfacecolor='None',markersize=10,markeredgewidth=2)
      j=j+0.009
     
show()
pcolor(som.distance_map().T) 
colorbar()
j=0.001
 
  
for i,x in enumerate(X):
    w=som.winner(x)
    if(  x[1]==4):
      plot(w[0]+j,
         w[1]+j,
        markers[3],
        markeredgecolor=colors[3],markerfacecolor='None',markersize=10,markeredgewidth=2)
      j=j+0.009
     
show()
pcolor(som.distance_map().T) 
colorbar()
j=0.001
 
  
for i,x in enumerate(X):
    w=som.winner(x)
    if(  x[1]==5):
      plot(w[0]+j,
         w[1]+j,
        markers[4],
        markeredgecolor=colors[4],markerfacecolor='None',markersize=10,markeredgewidth=2)
      j=j+0.009
      
show()
pcolor(som.distance_map().T) 
colorbar()
j=0.001
for i,x in enumerate(X):
    w=som.winner(x)
    if(  x[1]==6):
      plot(w[0]+j,
         w[1]+j,
        markers[5],
        markeredgecolor=colors[5],markerfacecolor='None',markersize=10,markeredgewidth=2)
      j=j+0.009       
show()
pcolor(som.distance_map().T) 
colorbar()
j=0.001
 
  
for i,x in enumerate(X):
    w=som.winner(x)
    if(  x[1]==7):
      plot(w[0]+j,
         w[1]+j,
        markers[6],
        markeredgecolor=colors[6],markerfacecolor='None',markersize=10,markeredgewidth=2)
      j=j+0.009
     
            
show()
pcolor(som.distance_map().T) 
colorbar()
j=0.001
 
  
for i,x in enumerate(X):
    w=som.winner(x)
    if(  x[1]==8):
      plot(w[0]+j,
         w[1]+j,
        markers[7],
        markeredgecolor=colors[7],markerfacecolor='None',markersize=10,markeredgewidth=2)
      j=j+0.009
    
show()
pcolor(som.distance_map().T) 
colorbar()
j=0.001
 
  
for i,x in enumerate(X):
    w=som.winner(x)
    if(  x[1]==9):
      plot(w[0]+j,
         w[1]+j,
        markers[8],
        markeredgecolor=colors[8],markerfacecolor='None',markersize=10,markeredgewidth=2)
      j=j+0.009
    
show()
pcolor(som.distance_map().T) 
colorbar()

j=0.001
 
  
for i,x in enumerate(X):
    w=som.winner(x)
    if(  x[1]==10):
      plot(w[0]+j,
         w[1]+j,
        markers[9],
        markeredgecolor=colors[9],markerfacecolor='None',markersize=10,markeredgewidth=2)
      j=j+0.009
     
show()
```
