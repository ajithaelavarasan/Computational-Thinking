```
#HEAT MAP OF CONTROL CATEGORY IN EACH GAME
import numpy as np; np.random.seed(0)
import seaborn as sns; sns.set_theme()
import pandas as pd
data = pd.read_csv('/dataset_after_data_aggregation.csv')
emp=[]
for x in data.Control:
  emp.append(x)
emp=np.array(emp).reshape(25,33)
ax = sns.heatmap(emp,square=True)
emp=[]

```
