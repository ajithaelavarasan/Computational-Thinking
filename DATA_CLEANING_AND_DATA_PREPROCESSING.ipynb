```python
import numpy
import pandas as pd
data = pd.read_csv('/home/ajitha/Documents/pro/datasetforbm.csv')
print(data)
output=[]
elements = data.TotalCommands
mean = numpy.mean(elements)
print(mean)
sd = numpy.std(elements)
print(sd)
for x in data.TotalCommands:
    if(x < mean - 2 * sd)or(x > mean + 2 * sd):
        if x not in output:
            output.append(x)            
data = data.set_index("TotalCommands")
for x in output:            
    data.drop(x,axis=0,inplace=True)
print(data)
data.to_csv(r'/home/ajitha/Documents/pro/dataset_after_datacleaning.csv')
```

```python
import numpy
import pandas as pd
data = pd.read_csv('/home/ajitha/Documents/pro/dataset_after_datacleaning.csv')
print(data)
df = pd.DataFrame(data,columns=['Motion','Looks','Sound','Events','Control','Sensing','Operators','Variables','Myblocks'])
SumCommands= df.sum(axis=1)
print (SumCommands)
data['SumCommands']=SumCommands
print(data)
binary=[]
for x in df.Motion:
    if x>0:
        binary.append(1)
    else:
        binary.append(0)
data['BinaryMotion']=pd.DataFrame(binary)
print(data)
binary=[]
for x in df.Looks:
    if x>0:
        binary.append(1)
    else:
        binary.append(0)
data['BinaryLooks']=pd.DataFrame(binary)
print(data)
binary=[]
for x in df.Sound:
    if x>0:
        binary.append(1)
    else:
        binary.append(0)
data['BinarySound']=pd.DataFrame(binary)
print(data)
binary=[]
for x in df.Events:
    if x>0:
        binary.append(1)
    else:
        binary.append(0)
data['BinaryEvents']=pd.DataFrame(binary)
print(data)
binary=[]
for x in df.Control:
    if x>0:
        binary.append(1)
    else:
        binary.append(0)
data['BinaryControl']=pd.DataFrame(binary)
print(data)
binary=[]
for x in df.Sensing:
    if x>0:
        binary.append(1)
    else:
        binary.append(0)
data['BinarySensing']=pd.DataFrame(binary)
print(data)
binary=[]
for x in df.Operators:
    if x>0:
        binary.append(1)
    else:
        binary.append(0)
data['BinaryOperators']=pd.DataFrame(binary)
print(data)
binary=[]
for x in df.Variables:
    if x>0:
        binary.append(1)
    else:
        binary.append(0)
data['BinaryVariables']=pd.DataFrame(binary)
print(data)
binary=[]
for x in df.Myblocks:
    if x>0:
        binary.append(1)
    else:
        binary.append(0)
data['BinaryMyblocks']=pd.DataFrame(binary)
print(data)
binary=[]
for x in df.Control:
    binary.append(x+2)
data['Complexity']=pd.DataFrame(binary)
print(data)
data.to_csv(r'/home/ajitha/Documents/pro/dataset_after_data_aggregation.csv')
    
```
