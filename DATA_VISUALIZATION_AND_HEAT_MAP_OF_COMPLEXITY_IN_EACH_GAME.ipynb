```
#DATA VISUALIZATION AND HEAT MAP OF COMPLEXITY IN EACH GAME 
import numpy
import pandas as pd
data = pd.read_csv('/dataset_after_data_aggregation.csv')
print(data)
output=[]
elements = data.Complexity
mean = numpy.mean(elements, axis=0)
print(mean)
sd = numpy.std(elements, axis=0)
print(sd)
for x in data.Complexity:
    if(x > mean - 2 * sd)and(x < mean + 2 * sd):
        if x not in output:
            output.append(x)
print(output)            
data = data.set_index("Complexity")
df=[]
for x in output:            
    data.drop(x,axis=0,inplace=True)
print(data)
#print(data)
data.to_csv(r'/dataset_not_limit_complexity.csv')
```
